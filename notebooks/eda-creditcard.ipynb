# Exploratory Data Analysis – Bank Credit Card Transactions

## Context
This dataset contains anonymized bank transaction data where most features have been transformed using PCA for privacy reasons.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

cc = pd.read_csv("../data/raw/creditcard.csv")
cc.head()
```

Features V1 to V28 are PCA-transformed variables.  
As a result, direct semantic interpretation is not possible, and pattern-based modeling is required.


```python
class_dist = cc['Class'].value_counts(normalize=True) * 100
class_dist
sns.barplot(x=class_dist.index, y=class_dist.values)
plt.title("Class Distribution – Credit Card Fraud")
plt.ylabel("Percentage")
plt.show()
```

The dataset exhibits extreme class imbalance, which is typical for banking fraud detection systems.


```python
sns.boxplot(x='Class', y='Amount', data=cc)
plt.title("Transaction Amount vs Fraud")
plt.show()
```

```python
sns.histplot(data=cc, x='Time', hue='Class', bins=50)
plt.title("Transaction Time vs Fraud")
plt.show()
```
